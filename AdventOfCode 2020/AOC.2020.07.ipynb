# part 1

```python
def fct(x):
    digits = ''.join(filter(lambda y: y.isdigit(), x))
    color = x[x.index(digits)+len(digits):].strip()
    if 'white' in color:
        print(f'"{color}", {x}')
    if digits == '':
        return None
    else:
        return (int(digits), color)
```

```python
rules = {}
m = 2
for line in data:
    els = map(lambda x: x.strip(), line.split('bag'))
    els = filter(lambda x: x != '.' and x != 's.', els)
    
    els = list(els)
    main = els[0]
    others = els[1:]
    
    others = map(fct, others)
    
    rules[main] = {el[1]: el[0] for el in list(others) if el}
```

```python
searched = 'shiny gold'
```

```python
def get(color, r=0):
    global used
    if color in used:
        return True
    if r > 8:
        return False
    
    if searched in rules[color]:
        used.append(color)
        print('+', end='')
        return True
    
    return any(get(sc, r+1) for sc in rules[color])
```

```python
total = 0
used = []
for i, k in enumerate(rules):
    if get(k):
        print('+', end='')
        total += 1
    else:
        print('-', end='')
    #print('\n====')
    #if i > 2:
    #    break
```

# Part 2

```python
waiting_list = {'shiny gold': 1}
```

```python
tot_by_round = {}
tot = 0
for i in range(100):
    if not waiting_list:
        print(f'Done in {i} turn')
        break
        
    pre_waiting = {}
    
    tot_by_round[i] = 0
    
    for name, amount in waiting_list.items():
        for k,v in rules[name].items():
            if not k in pre_waiting:
                pre_waiting[k] = 0
            pre_waiting[k] += v*amount
            tot += v*amount
            tot_by_round[i] += v*amount
            
    waiting_list = pre_waiting
    print('+', end='')
```
