```python
sc0 = np.array([list(map(int,el.split(','))) for el in aoc.data[0][1:]])
sc1 = np.array([list(map(int,el.split(','))) for el in aoc.data[1][1:]])
```

```python
sc0r = np.sort(sc0, axis=0)[0]
```

```python
sc0r
```

```python
rotations = []
for i, j, k in permutations(range(3), 3):
    sc = sc1[:, [i, j, k]]
    for l in range(2**3):
        ri = 1 if l%2 else -1
        rj = 1 if l//2%2 else -1
        rk = 1 if l//4%2 else -1
        scr = sc * (ri, rj, rk)
        scr.sort(axis=0)
        rotations.append((sc, scr - scr[0]))
```

```python
len(rotations)
```

```python
diffs = {}
for i, (real, relative) in enumerate(rotations):
    diffs[i] = [real_line for real_line, relative_line in zip(real, relative) if relative_line in sc0]
```
